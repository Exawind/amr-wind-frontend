# Plotting ABL profiles and statistics

```python
%%capture 

# Important header information
amrwindfedir = '../'
import sys, os
sys.path.insert(1, amrwindfedir)
os.chdir(amrwindfedir)  # Switch to this directory

# Load the libraries
import matplotlib.pyplot    as plt
import amrwind_frontend as amrwind

# Also ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Make all plots inline 
%matplotlib inline
```

```python
# Start the amrwind_frontend app and load the statistics file
app=amrwind.MyApp.init_nogui()
app.ABLpostpro_loadnetcdffile(ablfile='abl_statistics00000.nc')
```

### Plot the scalar time history

```python
# Plot the scalar time history 
plt.rc('font', size=14)
fig, ax = plt.subplots(figsize=(10,10), facecolor='w')
app.ABLpostpro_plotscalars(ax=ax, plotvars=['ustar'])
ax.set_ylabel('Ustar');
```

```python
# Plot the profiles
plt.rc('font', size=14)
fig, ax = plt.subplots(figsize=(10,10), facecolor='w')
app.ABLpostpro_plotprofiles(ax=ax, plotvars=['velocity', 'Uhoriz'], avgt=[500,600])
ax.set_ylim([0, 500]);
```
